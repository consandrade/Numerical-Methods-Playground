## Midpoint Summation

## Example 1:
Numerically integrate:
$$ y = x $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define your function f(x)
def f(x):
    return x

# Parameters
a = 0
b = 10
n = 1000
dx = (b - a) / n
x_values = np.linspace(a, b, n)  # Correctly define x_values

def Integrate(f):
    summation = 0
    Integral = []
    for i in range(n):
        summation += f(a + (i + 0.5) * dx) * dx
        Integral.append(summation)
    return Integral

Integral = Integrate(f)

# Plot Results
plt.plot(x_values, f(x_values), label="f(x) = x", linestyle="dashed")
plt.plot(x_values, Integral, label="Integral of f(x)")
plt.xlabel("x")
plt.ylabel("Function Value")
plt.title("Function vs. Its Integral")
plt.legend()
plt.show()
```
