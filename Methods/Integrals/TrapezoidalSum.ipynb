# Trapezoidal Sum

## Example 1:
$$ y = tan(x) $$

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return np.tan(x)

x_values = np.linspace(0, np.pi / 3, 1000)
a = 0
b = np.pi / 3
n = 1000
dx = (b - a) / n

def Integrate(f):
    summation = 0
    Integral = []
    for i in range(n):
        # Compute the integral at each step
        integral_value = (2 * summation + f(a) + f(b)) * dx / 2
        Integral.append(integral_value)
        summation += f(a + i * dx)  # Sum terms for the interior
    return Integral

Integral = Integrate(f)

plt.plot(x_values, f(x_values), label='f(x)')
plt.plot(x_values, Integral, label='Integral of f(x)')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
plt.show()
```
