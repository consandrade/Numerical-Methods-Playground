# Numerical Derivatives

## Example 1:
Numerically differentiate using the forward difference method: $$ y = cos(x) $$

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define your function f(x)
def f(x):
    return np.cos(x)

# params
x = np.linspace(0, 2 * np.pi)
dx = 0.01
dydx_list = []

# define algrithm of choice
def derivative(x, f):
    for i in x:
        dydx = (f(i + dx) - f(i)) / dx
        dydx_list.append(dydx)
    return dydx_list

dydx_list = derivative(x, f)

# plot results
plt.plot(x, f(x), label='cos(x)')
plt.plot(x, dydx_list, label='sin(x)')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```
