# Numerical Derivatives

## Finite Difference Methods
This code will show the implementations of the Finite Difference Methods on different functions.

## Example 1:
Numerically differentiate using the forward difference method: $$ y = cos(x) $$

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt

# Define your function f(x)
def f(x):
    return np.cos(x)

# params
x = np.linspace(0, 2 * np.pi) # x-values
dx = 0.01 # step size
dydx_list = [] # derivative list

# define algrithm of choice
def derivative(x, f):
    for i in x:
        dydx = (f(i + dx) - f(i)) / dx
        dydx_list.append(dydx)
    return dydx_list

dydx_list = derivative(x, f)

# plot results
plt.plot(x, f(x), label='cos(x)')
plt.plot(x, dydx_list, label='-sin(x)')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```

## Example 2:
Numerically differentiate using the central difference method:
$$ y = sin(x) $$

```python
def f(x):
    return np.sin(x)

x = np.linspace(0, 2 * np.pi)
dx = 0.01
dydx_list = []

def derivative(x, f):
    for i in x:
        dydx = (f(i + dx) - f(i - dx)) / (2 * dx)
        dydx_list.append(dydx)
    return dydx_list

dydx_list = derivative(x, f)

plt.plot(x, f(x), label='sin(x)')
plt.plot(x, dydx_list, label='cos(x)')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```

## Example 3:
Differentiate using the backwards difference method:
$$ y = e^x $$

```python
def f(x):
    return np.exp(x)

x = np.linspace(0, 20)
dx = 0.01
dydx_list = []

def derivative(x, f):
    for i in x:
        dydx = (f(i) - f(i - dx)) / dx
        dydx_list.append(dydx)
    return dydx_list

dydx_list = derivative(x, f)

plt.plot(x, f(x), label='e^x')
plt.plot(x, dydx_list, label = 'e^x')
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()

# Note, the derivative is plotting but it is hard to see at this resolution because the error is so low.
```
