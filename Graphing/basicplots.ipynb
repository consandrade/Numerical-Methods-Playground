## Basic plot using self initialized lists

```python
# import necessary libraries
import numpy as np # may not always be necessary
import matplotlib.pyplot as plt

# initialize lists/arrays of your x and y values
x = [1, 2, 3, 4, 5]
y = [1, 2, 3, 4, 5]

plt.plot(x, y) # plt.plot() ONLY takes lists and arrays and functions as arguments for what it will plot.
plt.xlabel('x - axis')
plt.ylabel('y - axis')
```

## Plotting multiple functions and using a legend

```python
# define functions to plot
def f(x):
    return np.cos(x)

def g(x):
    return np.sin(x)

x = np.linspace(0, 2 * np.pi) # np.linspace() creates a numpy array that can be used as an arg to plt.plot()

plt.plot(x, f(x), label='cos(x)')
plt.plot(x, g(x), label='sin(x)')
plt.xlabel('x - axis') # sets axis label based on string argument
plt.ylabel('y - axis') # sets axis label based on string argument
plt.legend() # plt.legend() adheres to the label you gave in your plt.plot args
```

## Setting the line color of your functions

```python
def f(x):
    return np.cos(x)

def g(x):
    return np.sin(x)

x = np.linspace(0, 2 * np.pi)

plt.plot(x, f(x), label='cos(x)', color='red') # color can be taken in any order if.f you put color =
plt.plot(x, g(x), label='sin(x)', color='green') # color can be taken in any order if.f you put color =
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.legend()
```
